## Use set index function to set index as patient number

```python
scrupdrs1 = scrupdrs1.set_index("PATNO")
```

```python
dataindex=[int(i) for i in scrupdrs1.index]
```

```python
p=[]
c=[]
s=[]
pro=[]
for i in patients["PATNO"]:
    p.append(int(i))
for  i in controls["PATNO"]:
    c.append(int(i))
for i in swedds["PATNO"]:
    s.append(int(i))
for i in prodormal["PATNO"]:
    pro.append(int(i))
```

```python
for i in p:
    if i in dataindex:
        scrupdrs1.loc[i,"DIAG"] = "PD"
```

```python
for i in c:
    if i in dataindex:
        scrupdrs1.loc[i,"DIAG"] = "HC"
```

```python
for i in pro:
    if i in dataindex:
        scrupdrs1.loc[i,"DIAG"] = "ProDorm"
```

```python
for i in s:
    if i in dataindex:
        scrupdrs1.loc[i,"DIAG"] = "Swedds"
```
